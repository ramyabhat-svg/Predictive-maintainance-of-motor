```python
import serial
import time

# Initialize serial communication with Arduino
arduino = serial.Serial('COM3', 9600)
time.sleep(2)  # Allow Arduino to reset

TEMP_THRESHOLD = 26.2  # Celsius
motor_state = True  # Motor starts ON

# Step 1: Turn motor ON initially
arduino.write(b'R')  # Command to start motor
print("Motor started")
time.sleep(4)  # Step 2: Wait before starting sensor check
print("Starting sensor monitoring...")

while True:
    try:
        data = arduino.readline().decode().strip()
        if data:
            parts = data.split(',')
            if len(parts) == 2:
                temperature = float(parts[0])
                vibration = parts[1].strip().upper()

                print(f"Temperature: {temperature}°C, Vibration: {vibration}")

                # Condition to stop motor
                if (temperature > TEMP_THRESHOLD or vibration == 'HIGH') and motor_state:
                    arduino.write(b'M')  # Turn Motor OFF
                    motor_state = False
                    print("Motor OFF due to high temperature or vibration")

                # Condition to start motor
                elif temperature <= TEMP_THRESHOLD and vibration == 'LOW' and not motor_state:
                    arduino.write(b'R')  # Turn Motor ON
                    motor_state = True
                    print("Motor ON")

            else:
                print("Invalid data format from Arduino:", data)

    except Exception as e:
        print(f"Error: {e}")
```
